```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("train.csv")
train
```

```python
train["NumOfProducts"].value_counts()
```

```python
data=train.drop(columns=["RowNumber","CustomerId","Surname"]) #拿掉較不重要的特徵
data=pd.get_dummies(data,columns=["Geography","Gender","NumOfProducts"]) #將類別型資料已one-hot encoding處理
data
```

```python
std=StandardScaler() #連續型以標準化處理
data[["CreditScore","Age","Tenure","Balance","EstimatedSalary"]]=std.fit_transform(data[["CreditScore","Age","Tenure","Balance","EstimatedSalary"]])
data
```

```python
y=data["Exited"].values #分出x跟y
x=data.drop(columns=["Exited"]).values
```

```python
test=pd.read_csv("test.csv") #以同樣方式處理測試集
test=test.drop(columns=["RowNumber","CustomerId","Surname"])
test=pd.get_dummies(test,columns=["Geography","Gender","NumOfProducts"])
test[["CreditScore","Age","Tenure","Balance","EstimatedSalary"]]=std.fit_transform(test[["CreditScore","Age","Tenure","Balance","EstimatedSalary"]])
test
```

## SVM 

### rbf kernel

```python
from sklearn.svm import SVC 
svm=SVC(C=0.8,gamma="auto",kernel="rbf") #調整參數並建模
svm.fit(x,y)
```

```python
y_pred=svm.predict(x) 
y_pred
print(accuracy_score(y_pred,y))
```

```python
test_pred=svm.predict(test.values) #對測試集進行預測
np.unique(test_pred)
```

```python
upload=pd.read_csv("upload.csv") 
upload["Exited"]=test_pred
upload["Exited"].value_counts()
```

```python
upload.to_csv("upload_svmrbf.csv") #將預測結果加入excel檔
```
